# Gender distribution

```python
plt.figure(figsize= (5,5))
ax= sns.countplot(data = df, x= "Gender")
ax.bar_label(ax.containers[0])
plt.title("Gender Distribution")
plt.show()
```

```python
#Analysis of the above question:-
# from the above chart we have analysed that:
#the number of females in the data is more than the number of males
```

```python
#  impact of parents on   student score
gb= df.groupby("ParentEduc").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gb)
```

```python
# plot heat map
sns.heatmap(gb)
plt.show()
```

```python
# plot heat map
plt.figure(figsize= (4,5))
sns.heatmap(gb, annot = True)
plt.title("Relationship between parent's education and student's score")
plt.show()
```

```python
#analysis of the above question:-
# from the above chart we have concluded that the education of the parents 
#have a good impact the score 
```

```python
gb1= df.groupby("ParentMaritalStatus").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
gb1
```

```python
# plot heat map
plt.figure(figsize= (4,4))
sns.heatmap(gb1, annot = True)
plt.title("Relationship between parent's Martial Status and student's score")
plt.show()
```

```python
# analysis of the above question:-
#from the given chart we have concluded that there is no /negligible
# impact on the student's score due to their parent's martial status
```

# Distribution of Ethnic Groups

```python
#Find the count and find pie chart 
# Counting the number of occurrences for each ethnic group
groupA = df.loc[(df['EthnicGroup'] == "group A")].shape[0]
groupB = df.loc[(df['EthnicGroup'] == "group B")].shape[0]
groupC = df.loc[(df['EthnicGroup'] == "group C")].shape[0]
groupD = df.loc[(df['EthnicGroup'] == "group D")].shape[0]
groupE = df.loc[(df['EthnicGroup'] == "group E")].shape[0]

# List of labels and values for the pie chart
l = ["group A", "group B", "group C", "group D", "group E"]
mylist = [groupA, groupB, groupC, groupD, groupE]

# Plotting the pie chart
plt.pie(mylist, labels=l, autopct="%1.2f%%")
plt.title("Distribution of Groups")
plt.show()
```
